### Lets try building a content based filter for user with user id 1

```python
item_features=items.drop(labels=['movie title','release date','video release date','IMDb URL','unknown'],axis=1)
item_features.head(6)
```

#### Cleaning the data
We will check to see of all the movies are assigned to a genre. If not, we have to remove them

```python
#calculate sum of all rows and normalize each cateogy for each item
row_sum=item_features.sum(axis=1)
row_sum[row_sum==0]
```

```python
item_features.drop(item_features.index[266],inplace=True)
item_features.drop(item_features.index[1371],inplace=True)
```

```python
print('No of movies with id 267 or 1327 =',len(ratings[(ratings['movie id']==267)|(ratings['movie id']==1327)]))
ratings=ratings[(ratings['movie id']!=267)&(ratings['movie id']!=1327)]
print(len(ratings))
```

```python
#now we have deleted the two movies
row_sum=item_features.sum(axis=1)
row_sum[row_sum==0]
```

#### Lets normalize the item data

```python
#now we can normalize the features for each item(movie)
print(' 1st row before normalization: \n',item_features.iloc[1])
item_features=item_features.divide(np.sqrt(np.array(row_sum)),axis=0)
print('\n  1st row after normalization: \n',item_features.iloc[1])
```

```python
item_features[:1].sum(axis=0)
```

```python
#calculate DF and IDF for each feature for the items

#DF is number of movies in each genre OR frequency of each genre in all collection of movies. 
#we calculate this by taking the sum along the columns
column_sum=item_features.sum(axis=0)
print(column_sum)

#IDF is estimates as inverse log of DF
IDF=list(np.log10(len(item_features)/column_sum))
IDF
```

From the above caculations, we observe that the DF for drama and comedy are the highest as also previously observed in the bar plot. The IDF values are invesely related to DF, so drama and comedy will show the lowest IDF values. This allows us to counter the bias in recommendations, without which the genre 'Drama' or 'Comedy' would be assigned higher weights just because of their high frequency in the dataset.   

#### The dot product of movie feature vectors and IDF vectors gives us the weighted scores of each movie.

As seen below in the case of movie id 1, Animation, Children's and Comedy have decreasing weighted scores. This is becasue there are only 42 Animation movies compared to 122 and 505 Children's and Comedy movies respectively. Hence, the Animation genre gets a higher rating. 

```python
#lets calculate weighted score of features for each item by taking dot product of feature vector for each item with the IDF vector
item_features=item_features.multiply(IDF,axis=1)
item_features.iloc[0]
```

#### Now that we have the item vector for all the movies, we now need the user profile vector. 
Lets estimate that for **User 1** by using the movies he/she has rated. As seen below user 1 has rated 272 movies.

```python
#movies rated by user 1
user1_ratings=ratings[ratings['user_id']==1]
print(len(user1_ratings))
user1_ratings.set_index('movie id',inplace=True)
user1_ratings.sort_index(inplace=True)
user1_ratings
```

First, to make the user profile vector, we need a vector with the users ratings for the movies he/she has rated, and zeros for movies that they have not.

```python
user1_vector=np.zeros(len(item_features))
index=np.array(user1_ratings.index.values)-1
user1_vector[index]=user1_ratings['rating']

user1_preferences=[]

for column in list(item_features.columns.values):
    item_vector=np.array(item_features[column])
    user1_preferences.append(np.dot(item_vector,user1_vector))

user1_df=pd.DataFrame(user1_preferences,list(item_features.columns.values))
user1_df.rename(index=str, columns={0: "score"},inplace=True)
user1_df.sort_values(by='score',ascending=False,inplace=True)
user1_df
```

**This shows that User 1 likes Action, Sci-Fi, Comedy and Drama as top 4 liked genres.** 
Now, lets check the ratings User 1 has given for the movies he's rated and their respective genres. We can do this by joining the ratings and item_features dataframes using inner join. 

```python
#check
#user1_ratings.sort_values(by='rating',ascending=False,inplace=True)
user1_ratings=user1_ratings.join(item_features,how='inner')
user1_ratings
column_sum=user1_ratings.sum(axis=0).sort_values(ascending=False)
column_sum
```

And as estimated, User 1 behavior shows the highest ratings given to movies in Action, Comedy, Sci-Fi and Drama. Sci  Fi is below Comedy in this brute-force method-of-sum-estimation, but since there are only 101 sci-fi movies compared to 505 comedy movies, the actual scores shows sci-fi above comedy.

### Predict User 1 Ratings for all movies
#### Now we can predict User 1 behavior for all the movies. 
We can do this by computing dot product between the weighted scores for each movie and the user profile vector (user 1 here). This gives a probability that the user will like a particular movie.


```python
item_features.head()
```

```python
user1_preferences
```

```python
user1_prediction_scores=item_features.dot(user1_preferences)
user1_prediction_scores.sort_values(ascending=False).head(5)
```

### Limitations of Content-based Recommendation system
1) A content based recommedation system works great for some cases like movies, since a user who prefers action or comedy would prefer to watch more movies of the same genres. However, in cases like produc recommendations like in Amazon, a user who buys shoes, would be less likely to buy more shoes.
Hence, these use cases requries a different approach that takes into account multiple user's behavior.

2) Another major drawback is described below. This method is basically just estimating cosine similarity based on the dot product of the movie genre vectors and the user's rating profile vector. In reality, this is a more a measure of the movies the user will watch rather than which the user will like.

3) A 5-star rating level is an approximation at best. Different people have different rating habits. As seen below, the ratings for 4 and 5 stars is much more than 1 and 2 stars. So the distribution in ratings isnt uniform.

```python
sns.set_style('whitegrid')
sns.set(font_scale=1.5)

# Display distribution of rating
sns.distplot(ratings['rating'].fillna(ratings['rating'].median()))
```
